```python
from transformers import ViTImageProcessor, ViTForImageClassification
from PIL import Image
import requests
```

# Vision Transformer (ViT)

Transformer Encoder: ViT maakt gebruik van een standaard transformer-encoder bestaande uit meerdere lagen. Elke laag bestaat typisch uit een multi-head zelfaandachtsmechanisme en doorstuur-neurale netwerken. Het zelfaandachtsmechanisme stelt het model in staat om aandacht te schenken aan verschillende patches en relevante informatie vast te leggen.

Dingen die ViT uniek maken in vergelijking met een CNN model zijn:

De patch-gebaseerde benadering van ViT stelt het model in staat om afbeeldingen op te splitsen in kleinere patches en deze te behandelen als sequenties van vectoren. Dit maakt het mogelijk om transformer-architecturen toe te passen op afbeeldingsgegevens, waarbij het zelfaandachtsmechanisme wordt gebruikt om interacties tussen de verschillende patches vast te leggen.

Bovendien maakt ViT gebruik van het zelfaandachtsmechanisme om globale afhankelijkheden te begrijpen en vast te leggen in de hele afbeelding. Het model kan aandacht schenken aan alle patches en relaties begrijpen tussen patches die ver van elkaar verwijderd zijn. Door globale contextuele informatie vast te leggen, kan ViT belangrijke informatie gebruiken bij het maken van voorspellingen.

Samengevat, ViT onderscheidt zich van traditionele CNN-modellen door zijn patch-gebaseerde aanpak en het vermogen om globale afhankelijkheden vast te leggen. Dit stelt het model in staat om beter om te gaan met afbeeldingsgegevens en belangrijke contextuele informatie te gebruiken voor taken zoals beeldclassificatie.

```python

image = Image.open("../datasets/dieren/kat.jpg")

display(image)
```

```python
processor = ViTImageProcessor.from_pretrained('google/vit-base-patch16-224')
model = ViTForImageClassification.from_pretrained('google/vit-base-patch16-224')

```

```python
inputs = processor(images=image, return_tensors="pt")
outputs = model(**inputs)
logits = outputs.logits
# model predicts one of the 1000 ImageNet classes
predicted_class_idx = logits.argmax(-1).item()
print("Predicted class:", model.config.id2label[predicted_class_idx])
```
