## Eerste keer trainen
Deze code traint een Q-learning agent op de "FrozenLake-v1" omgeving met behulp van het Q-learning algoritme. Het bijwerken van de Q-tabel gebeurt op basis van de waargenomen beloningen en overgangen. Na het trainen wordt het uiteindelijke resultaat van de omgeving weergegeven.

```python
import numpy as np
import gym
import matplotlib.pyplot as plt

episodes = 1000
alpha = 0.1
gamma = 0.6
qtable = np.zeros((16, 4))
outcomes = []

environment = gym.make("FrozenLake-v1", render_mode='human')
environment.reset()
environment.render()



for i in range(episodes):
    print(f"Episode: {i+1}")
    test = environment.reset()
    state = test[0]
    done = False
    outcomes.append("Failure")

    while not done:
        if np.max(qtable[state]) > 0:
            action = np.argmax(qtable[state])
        else:
            action = environment.action_space.sample()

        new_state, reward, done, info, j = environment.step(action)

        qtable[state, action] = qtable[state, action] + alpha * (reward + gamma * np.max(qtable[new_state]) - qtable[state, action])

        state = new_state
        print(f'Reward = {reward}')

# Display the results (reward and map)
environment.render()

```

## testen

```python
success_rates = []
success_count = 0

for i in range(episodes):
    print(f"Episode: {i+1}")
    test = environment.reset()
    state = test[0]
    done = False

    while not done:
        action = np.argmax(qtable[state])
        new_state, reward, done, info, _ = environment.step(action)
        state = new_state

    if reward == 1:
        success_count += 1

    success_rate = success_count / (i + 1)
    success_rates.append(success_rate)

# Plotting the success rate
print(success_count)
plt.figure(figsize=(8, 6))
plt.plot(range(1, episodes + 1), success_rates)
plt.xlabel('Episode')
plt.ylabel('Success Rate')
plt.title('Success Rate over Episodes')
plt.show()
```

## Tweede keer trainen maar met epsilon

Door epsilon langzaam te verminderen, neemt de exploratie af en neemt de exploitatie toe naarmate de agent betere Q-waarden leert en meer gerichte acties kan nemen op basis van die waarden. Hierdoor kan de agent in de loop van de tijd betere prestaties behalen en optimalere acties uitvoeren in de omgeving.



```python
import numpy as np
import gym
import matplotlib.pyplot as plt

episodes = 1000
alpha = 0.1
gamma = 0.6
epsilon = 1.0
epsilon_decay = 0.001
epsilon_min = 0.01
qtable = np.zeros((16, 4))
successes = []
outcomes = []

environment = gym.make("FrozenLake-v1", render_mode='human')

for i in range(episodes):
    print(f"Episode: {i+1}")
    state = environment.reset()[0]
    done = False
    outcomes.append("Failure")
    success = 0

    while not done:
        if np.random.uniform(0, 1) < epsilon:
            action = environment.action_space.sample()  # Exploration
        else:
            action = np.argmax(qtable[state])  # Exploitation

        new_state, reward, done, info, j = environment.step(action)

        qtable[state, action] += alpha * (reward + gamma * np.max(qtable[new_state]) - qtable[state, action])

        state = new_state
        success += reward  # Track cumulative rewards for success rate calculation

    successes.append(success)  # Track the cumulative rewards for each episode

    epsilon = max(epsilon_min, epsilon - epsilon_decay)  # Decay epsilon over time

# Plot the success rate over episodes
success_rate = np.cumsum(successes) / np.arange(1, episodes + 1)
plt.plot(success_rate)
plt.xlabel('Episodes')
plt.ylabel('Success Rate')
plt.title('Success Rate over Episodes')
plt.show()

# Display the results (reward and map)
environment.render()
```

## Testen

```python
successes_test = []
outcomes_test = []

for i in range(100):
    print(f"Episode: {i+1}")
    state = environment.reset()[0]
    done = False
    outcomes_test.append("Failure")
    success = 0

    while not done:
        action = np.argmax(qtable[state])

        new_state, reward, done, info, j = environment.step(action)

        state = new_state
        success += reward

    successes_test.append(success)


```

```python
success_rate_test = np.cumsum(successes_test) / np.arange(1, 100 + 1)
print("Test Success Rate:", success_rate_test[-1])

# Plot the test success rate
plt.plot(success_rate_test)
plt.xlabel('Episodes')
plt.ylabel('Success Rate')
plt.title('Test Success Rate over Episodes')
plt.show()
```
