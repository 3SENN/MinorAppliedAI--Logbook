## 3. Modellen
Op basis van de dataset met studentencijfers heb ik gekozen voor het Linear Regression-model,  K-Nearest Neighbors Regressor met 5 buren (KNN) en de Decision Tree Regressor om nauwkeurige voorspellingen te maken. Ik heb gekozen voor het Linear Regression-model omdat het een eenvoudig en veelgebruikt model is voor regressieproblemen. Het kan inzicht bieden in de relatie tussen de onafhankelijke variabelen (zoals studietijd, aanwezigheid en eerdere cijfers) en de afhankelijke variabele (de cijfers). Met behulp van dit model kan ik de relatie tussen de verschillende factoren en de cijfers begrijpen. Bijvoorbeeld, ik kan bepalen hoe veranderingen in studietijd of aanwezigheid van invloed zijn op de cijfers. Het Linear Regression-model geeft ook coëfficiënten voor elke inputvariabele, waardoor ik de relatieve belangrijkheid van verschillende factoren in het voorspellen van de cijfers kan begrijpen.

In het geval van het voorspellen van studentencijfers kan de KNN-regressor nuttig zijn omdat het rekening houdt met vergelijkbare studenten op basis van hun kenmerken, zoals studietijd, aanwezigheid en eerdere cijfers. Door te kiezen voor 5 buren, overweeg ik een redelijk aantal vergelijkbare studenten bij het maken van voorspellingen. De KNN-regressor kan effectief lokale patronen en relaties in de data vastleggen. Bijvoorbeeld, als een student vergelijkbare kenmerken heeft als zijn/haar 5 dichtstbijzijnde buren, kan ik verwachten dat de voorspelde cijfers van deze buren een goede indicatie geven van het mogelijke cijfer van de student.

Daarnaast heb ik gekozen voor de Decision Tree Regressor. Deze regressor kan omgaan met niet-lineaire relaties in de data. In het geval van het voorspellen van studentencijfers kunnen er complexe interacties en niet-lineaire afhankelijkheden zijn tussen verschillende factoren (bijvoorbeeld de relatie tussen studietijd, aanwezigheid en cijfers is mogelijk niet lineair). De Decision Tree Regressor kan dergelijke niet-lineaire relaties vastleggen door een boomachtig model van beslissingen te construeren op basis van de inputvariabelen. Het geeft een duidelijke visualisatie van het besluitvormingsproces, wat waardevol kan zijn bij het begrijpen van de factoren die van invloed zijn op de cijfers. Bijvoorbeeld, ik kan specifieke drempelwaarden voor studietijd of aanwezigheid identificeren die leiden tot verschillende voorspellingen van de cijfers.

Door deze drie algoritmen te combineren, probeer ik hun verschillende sterke punten te benutten om nauwkeurige voorspellingen voor studentencijfers te maken. Het Linear Regression-model legt de algemene trend en de impact van individuele kenmerken vast, de Decision Tree Regressor kan complexe niet-lineaire relaties vastleggen en de beslissingsregels visualiseren, en de KNN-regressor kan profiteren van de vergelijkbaarheid van studenten om lokale patronen in de data te begrijpen.

## Prepare Data

```python
df = data
df = df.drop(['famrel','absences','Mjob','Fjob','school','studytime','health'], axis=1)

# Replaces the string values 'F' and 'M' in the 'Sex' column with 0 and 1
df['sex'].replace({'F': 0, 'M': 1},inplace = True)
df.rename(columns={'sex': 'Male'}, inplace=True)

df['Pstatus'].replace({'T': 0, 'A': 1},inplace = True)
df.rename(columns={'Pstatus': 'ParentsApart'}, inplace=True)

df['schoolsup'].replace({'no': 0, 'yes': 1},inplace = True)
df['famsup'].replace({'no': 0, 'yes': 1},inplace = True)
df['paid'].replace({'no': 0, 'yes': 1},inplace = True)
df['activities'].replace({'no': 0, 'yes': 1},inplace = True)
df['nursery'].replace({'no': 0, 'yes': 1},inplace = True)
df['higher'].replace({'no': 0, 'yes': 1},inplace = True)
df['internet'].replace({'no': 0, 'yes': 1},inplace = True)
df['romantic'].replace({'no': 0, 'yes': 1},inplace = True)

df['address'].replace({'R': 0, 'U': 1},inplace = True)
df.rename(columns={'address': 'addressUrban'}, inplace=True)

df['famsize'].replace({'LE3': 0, 'GT3': 1},inplace = True)
df.rename(columns={'famsize': 'famsizeGR3'}, inplace=True)

# Onehot encode daily alcohol
onehot = pd.get_dummies(df['Dalc'], prefix='Dalc')
df = pd.concat([df, onehot], axis=1)
df = df.drop('Dalc', axis=1)

# Onehot encode Reason
onehot = pd.get_dummies(df['reason'], prefix='reason')
df = pd.concat([df, onehot], axis=1)
df = df.drop('reason', axis=1)

# Onehot encode guardian
onehot = pd.get_dummies(df['guardian'], prefix='guardian')
df = pd.concat([df, onehot], axis=1)
df = df.drop('guardian', axis=1)

# Calculate the average value of 'age' excluding outliers
average_age = np.mean(df.loc[~outliers_age, 'age'])

# Replace outliers with the average age
df.loc[outliers_age, 'age'] = average_age
```

## T-test
Het is handig om een t-test uit te voeren in dit scenario omdat het je in staat stelt om te beoordelen of er een statistisch significant verschil is tussen de gemiddelde waarden van twee onafhankelijke groepen, in dit geval de 'Fedu' en 'Medu' variabelen. Als de p-waarde kleiner is dan een vooraf bepaald significantieniveau (bijvoorbeeld 0.05), kunnen we concluderen dat er voldoende bewijs is om aan te nemen dat er een echt verschil is tussen de gemiddelden van de twee variabelen.

Waarom ik dit wil weten: Ik vraag mij af of er een verschil is in Fedu en Medu in educatie van leerlingen. Hierdoor kan ik kijken of de educatie van de ouders uiteindelijk ook een invloed heeft op de behaalde cijfers van de studenten.

```python
from scipy.stats import ttest_ind
# uitvoeren van de t-test
t_value, p_value = ttest_ind(df['Fedu'], df['Medu'])

# tonen van de t-test resultaten
print("t-waarde:", t_value)
print("p-waarde:", p_value)
```

Het verschil tussen de gemiddelde waarden van de twee steekproeven is significant. (Slechts 0,34% kans dat het waargenomen verschil tussen de twee steekproeven door toeval wordt veroorzaakt.) Gemiddelde van Fedu is 2.9 kleiner dan Medu.

## PCA
De prestaties van een regressiemodel worden vergelijken wanneer alle beschikbare features worden gebruikt versus wanneer de features worden verminderd met behulp van Principal Component Analysis (PCA).

Door de prestaties van het model met alle features te vergelijken met het model dat is gebaseerd op de gereduceerde features door PCA, kun je beoordelen of het gebruik van een subset van features (in dit geval de PCA-componenten) voldoende is om de voorspelkracht van het model te behouden. Dit kan handig zijn om de dimensionaliteit van de dataset te verminderen en de rekentijd te verkorten, terwijl je toch redelijk accurate voorspellingen kunt doen

```python
# schaal de features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# voer een PCA uit en reduceer de features
pca = PCA(n_components=5)
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)

# voorspel de uitkomst met alle features
model_all = LinearRegression()
model_all.fit(X_train_scaled, y_train)
y_pred_all = model_all.predict(X_test_scaled)
r2_all = r2_score(y_test, y_pred_all)
mse_all = mean_squared_error(y_test, y_pred_all)

# voorspel de uitkomst met een aantal componenten
model_pca = LinearRegression()
model_pca.fit(X_train_pca, y_train)
y_pred_pca = model_pca.predict(X_test_pca)
r2_pca = r2_score(y_test, y_pred_pca)
mse_pca = mean_squared_error(y_test, y_pred_pca)

# vergelijk de uitkomsten en trek conclusies
print("R2 score met alle features: {:.2f}".format(r2_all))
print("Mean Squared Error met alle features: {:.2f}".format(mse_all))
print("R2 score met PCA: {:.2f}".format(r2_pca))
print("Mean Squared Error met PCA: {:.2f}".format(mse_pca))
```

## PCA Output analyse
Wat opvalt bij de output van de PCA: Is dat de R2 score met alle features een hogere R2 score heeft en een lagere MSE. Dit kan erop wijzen dat het model met alle features beter presteert dan het model met gereduceerde features.

Want als de MSE hoger is betekent dat de voorspellingen van het model gemiddeld gezien verder van de werkelijke waarden af liggen.

# XAI extensie LIME

LIME is een techniek die probeert de voorspellingen van een black-box machine learning-model te verklaren op een lokaal niveau.
Met behulp van LIME genereert de explain_instance-methode een uitleg die de bijdrage van elke functie aan de voorspelling voor het specifieke voorbeeld laat zien.

- Feature: Een feature (kenmerk) verwijst naar een specifieke eigenschap of variabele van een voorbeeld in de dataset. Bijvoorbeeld, als we een classificatiemodel hebben dat auto's en fietsen onderscheidt, kunnen de kenmerken onder andere het merk, de kleur, het gewicht, de snelheid, enzovoort zijn. Elk kenmerk heeft een bijbehorende waarde.

- Value: De waarde van een kenmerk is de specifieke waarde die wordt geassocieerd met dat kenmerk voor een bepaald voorbeeld. Bijvoorbeeld, als het kenmerk "merk" is en het voorbeeld een auto is, kan de waarde "Toyota" zijn. Als het kenmerk "gewicht" is, kan de waarde bijvoorbeeld 1500 kg zijn.

In het geval van LIME, wordt een specifiek voorbeeld (example) uit de dataset geanalyseerd om te begrijpen welke kenmerken en waarden hebben bijgedragen aan de voorspelling van het model.

De "prediction probabilities" verwijzen naar de voorspelde kansen of waarschijnlijkheden van de verschillende klassen die het model kan voorspellen. De voorspelde kansen geven aan hoe zeker het model is over de toewijzing van een voorbeeld aan een specifieke klasse. In LIME worden deze voorspelde kansen gebruikt om de uitleg te genereren en te begrijpen hoe de kenmerken bijdragen aan de voorspelling.

![LIME](https://gitlab.fdmci.hva.nl/chintss/minor-logboek-aai-2/-/raw/main/machinelearning/Screenshot_2023-06-09_at_22.35.59.png)

```python
from lime import lime_tabular
example = X_test[0]
explainer = lime_tabular.LimeTabularExplainer(X_train, feature_names=feature_names, discretize_continuous=True)
explanation = explainer.explain_instance(example, rf_classifier.predict_proba, num_features=5)
explanation.show_in_notebook()  # Toon de verklaring in een Jupyter Notebook
```
